```python
# A scatter plot is a type of data visualization that displays individual data points as dots on a 2D plane. Each dot represents a single observation with its position determined by the values of two variables, one plotted on the x-axis and the other on the y-axis.

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

# Display the plot
plt.show()
```

```python
# The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange plots in a grid format, specifying the number of rows, number of columns, and the position of each plot within the grid. 

import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.subplot(2, 2, 1)  
plt.plot(x1, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2) 
plt.plot(x2, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3) 
plt.plot(x3, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)  
plt.plot(x4, y4)
plt.title('Line Plot 4')
plt.tight_layout()
plt.show()
```

```python
#  bar plot also known as a bar chart or bar graph, is a data visualization tool that uses rectangular bars to represent data. Each bar's length corresponds to the value of a specific category or variable, making it easy to compare and display the relative sizes of different data points.

import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits (Bar Plot)')
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')
plt.show()
```

```python
# A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It provides a visual summary of the data's median, quartiles, and any potential outliers. 

import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot for box1 and box2')

plt.show()
```
